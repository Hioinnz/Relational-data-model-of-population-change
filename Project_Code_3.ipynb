```r
unique_countries <- Population %>%
  filter(Region_Name == "Europe") %>%
  distinct(Country_Name) %>%
  count()

Highest_latitude <- Population %>%
  filter(Region_Name == "Europe") %>%
  arrange(desc(Latitude)) %>%
  pull(Latitude) %>%
  first()

Lowest_latitude <- Population %>%
  filter(Region_Name == "Europe") %>%
  arrange(Latitude) %>%
  pull(Latitude) %>%
  first()

percentiles <- c(25, 50, 75)
values <- length(percentiles)
for (i in 1:length(percentiles)) {
        values[i] <- Lowest_latitude + (percentiles[i] / 100) * (Highest_latitude - Lowest_latitude)
}


print(paste("Number of European Countries in dataset :", unique_countries))
print("Latitude in Europe")
print(paste("Highest :", Highest_latitude))
print(paste("75% :", values[3]))
print(paste("50% :", values[2]))
print(paste("25% :", values[1]))
print(paste("Lowest :", Lowest_latitude))
```

```r
EU_latitude <- function(Latitude){

    result <- vector(mode = "character", length = length(Latitude))
    for(i in 1:length(Latitude)) {
        testing_element <- Latitude[i]
        if (testing_element >= 35.83 & testing_element < 43.1225) {
            result[i] <- "Low Latitude"
        } else if (testing_element >= 43.1225 & testing_element < 50.415){
            result[i] <- "Middle-Low Latitude"
        } else if (testing_element >= 50.415 & testing_element < 57.7075){
            result[i] <- "Middle-High Latitude"
        } else if (testing_element >= 57.7075 & testing_element <= 65){
            result[i] <- "High Latitude"
        }
    }
    return(result)
}
```

### comment

Europe is the only continent which has a population decease. We were curious that if all European countries contribute to it or not. This time we divided the European countries into 4 different group depending on it's latitude. We want to see if the geographic factors also would have an impact on population. (Europe is the most appropriate region to use this method because we have only one latitude and longitude for each country. In Europe there is not many big countries compared to other continent where china or India has too big land for single combination of latitude and longitude) To do this, we defined a function called `EU_latitude` to wrangle the data for visualisation.

```r
Europe_Lat <- Population %>%
    filter(Region_Name == "Europe") %>%
    na.omit() %>%
    mutate(Lat = Latitude %>% EU_latitude()) %>%
    group_by(Lat, Year) %>%
    summarise(Mean_Population = mean(Population)) %>%
    ggplot(aes(x = Year, y = Mean_Population, group = Lat, color = Lat)) +
    geom_line(lwd=1, show.legend = TRUE) +
    ggtitle("Population changes in Europe depending on Latitude") +
    xlab("Year") +
    ylab("Growth") +
    labs(color = "Latitude Level")

Europe_Lat
```

### comment

As mentioned above, we divied the European cointuries into 4 different group. To do this, we used `EU_latitude` function and `mutate()` function to make a new column called **Lat** with 4 latitude groups. Some **Latitude** values are missing (mentioned in notebook2), so we needed to use `na.omit()` function to delete the missing value observations. Then, the data grouped by **Year** and **Lat** and summarised with mean of population of each group. ggplot line graph was used for visualisation and the line colors are different upto it's latitude level.

If you see the graph, not all the graph is going down. Only red(High Latitude) and purple(Middle-Low Latitude) lines have a dropping point after 1990. Here is the time when Europe's population stopped increasing. This means not all European countries are contributing to a decline in population. The geographic factors would not affect to population decease directly but we showed that we can divide the countries into not only just continents but also geographic factors and it actually gave some meaningful insights.
In further research, we can compare those countries in different group and identify what kind of factors could make the difference between those 4 different latitude groups.

```r
Economic <- read_csv("Economic_factors_Clean.csv")

Economic
```

```r
Economic %>%
    skimr::skim()
```

### comment

Read `Economic_factors_Clean.csv` files saved in notebook 1. It has 2 character type variables and 4 numeric type variables. We can join this table to `population` table to see the economic impact on demographic figures.

```r
pop_eco <- Population %>%
    filter(Year > 1990) %>%
    left_join(Economic %>% select(-Country_Name),
              by = c("Country_Code", "Year"))

pop_eco
```

### comment

Joined `Population` and `Economic` tables by **Country_Code** and **Year** to visualise them together.

```r
library("ggplot2")                     
library("GGally")


pop_eco_pairs <- ggpairs(pop_eco, columns = c(4, 5, 6, 9, 10, 11), ggplot2::aes(colour=Region_Name))

pop_eco_pairs
```

### comment

We draw the pair plot for `pop_eco` data. The most noticeable points are two green lines which are set part to other points. We are assuming they are China and India because they are the only countries that have more than billion population. It looks like the most related economic factors to population is **GDP**. **GDP** and **Population** both increased over times. Also, the relationship between population and GDP is positively correlated across the continents. However, it is hard to tell that **GDP** has a direct causality to population because Americas' GDP is the highest in most of times but its population has never been the biggest. In addition, the GDP in China is growing faster than India but the slope of population growth in India is more steep than China. (We can see this closely in the next graph)

It looks like **Unemployment_rate** is related to population as well. Europe has the highest points of **Unemployment_rate**, but the continents where the population is growing steadily such as Asia and Americas have low **Unemployment_rate**. 

```r
China_India_eco <- pop_eco %>% 
    filter(Country_Name %in% c("China", "India")) %>%
    select(c(2, 4, 5, 6, 9, 10)) %>%
    mutate(Population = scale(Population),
           GDP = scale(GDP),
           Unemployment_rate = scale(Unemployment_rate),
           Population_Growth_Rate = scale(Population_Growth_Rate)) %>%
    gather(key = Factor,
         value = Value,
         -c(Country_Name, Year)) %>%
    ggplot(aes(x = Year, y = Value, group = Factor, color = Factor)) +
    geom_line(lwd = 1, show.legend = TRUE) +
    ggtitle("Economic Factors changes over times in China and India") +
    xlab("Year") + 
    ylab("Value (Standardised)") + 
    labs(color = "Factor")  + 
    facet_wrap(~ Country_Name)

China_India_eco    
```

### comment

To see China and India closely, we filtered the `pop_eco` to leave only China and India observations. As all the economic factors have different scales, it is hard to put them into one figure. Therefore, we standardised the values because the trends are more important than actual values. For visualisation, we needed to make the data table into long format one again, so the table have four features **Country_Name**, **Year**, **Factor**, and **Value**. By having the factors in one column, we can visualise the factors having different colors. Then, we used ggplot with wrangled table as usual and divided it into two graph depending on the **Country_Name** which only has either China and India. 

You can see what I explained above. The slope of GDP in China is much more steep than India, but the slope of Population in India is more steep than China. Based on this, we can assume that GDP (economic factors) is sufficient condition for population growth but not necessary conditon.

```r
Unemployment_box <- pop_eco %>%
    ggplot(aes(x = Region_Name, y = Unemployment_rate)) +
    geom_boxplot(outlier.colour = "hotpink") +
    ggtitle("Unemployment Rate of each Continent") +
    theme(plot.title = element_text(hjust = 0.5)) +
    ylab("Unemployment Rate") +
    geom_jitter(position = position_jitter(width=0.1, height =0 ), alpha =1/4)

Unemployment_box
```

### comment

To see **Unemployment_rate** closely, we polt the boxplot. Compared to others, Europe definitely has the most outliers at upside of the box. This can be a reason that Europe's population growth rate is low compared to others.

```r
pop_soc_pairs <- ggpairs(enriched_pop, columns = c(4, 5, 6, 12, 13, 14), ggplot2::aes(colour=Region_Name))

pop_soc_pairs
```

### comment

We draw the pair plot for `pop_soc` data. The noticeable features are **Life_expectancy** is increasing and **Birth_rate** is decreasing over times in all regions. Americas region has very high net migraion counts compared to other continents. The values are fluctuate every year but it has a increasing trend. Europe has very low **Birth_rate** and Africa still has high **Birth_rate**. The population in Africa continent was growing so fast but didn't not find any prominent evidence from economic factors, but it has the highest birth rate. The reason of the growing population in Africa can be just high birth rate because they also has lower life expectancy compared to others. This it quite opposite to Americas. The population is growing in Americas but their birth rate is very low and life expentancy is high. Even if Americas have low birth rate they overcome it with migration and long life expectancy. About Europe, like mentioned above, birth rate is very low but life expectancy is pretty high like Americas. However, they don't have many migration like Americas, so their population is not growing like Americas. 

```r
LE_BR <- enriched_pop %>%
    group_by(Year) %>%
    summarise(mean_birth_rate = mean(Birth_rate),
              mean_life_expectancy = mean(Life_expectancy),
              mean_population_growth_rate = mean(Population_Growth_Rate)) %>%
    ggplot(mapping = aes(x = Year)) +
    geom_line(aes(y = mean_birth_rate, color = "Mean Birth Rate"), lwd = 1) +
    geom_line(aes(y = mean_life_expectancy, color = "Mean Life Expectancy"), lwd = 1) +
    geom_line(aes(y = mean_population_growth_rate, color = "Mean Population Growth Rate"), lwd = 1) +
    scale_color_manual(values = c("Mean Birth Rate" = "blue", "Mean Life Expectancy" = "red", "Mean Population Growth Rate" = "hotpink")) +
    ggtitle("Life Expectancy and Birth Rate Over Times in the world") +
    xlab("Year") +
    ylab("Values") +
    labs(color = "Variable")

LE_BR
```

### comment

Overall, the population in the world are growing not because people have babies but becuase people don't die. The population growth rate is very stable becuase the gap between two variables are getting bigger at quite similar speed.
